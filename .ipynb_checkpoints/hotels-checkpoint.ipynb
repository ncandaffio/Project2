```python
#Import Dependancies
import pandas as pd
from Keys import ZIP
import requests
import pymongo
```

```python
#Use Pandas read_csv function to extract the data into a dataframe
csv = 'Data/7282_1.csv'
hotels_df = pd.read_csv(csv)
hotels_df = hotels_df.loc[hotels_df['country'] == 'US']
hotels_df.head()
```

```python
#Create a dataframe that is grouped by the name, returning the mean
hotels_mean = hotels_df.groupby(['name']).mean()
hotels_mean.head()
```

```python
#Create a dataframe that is grouped by the name, returning the max to extract the non-numeric data
hotels_max = hotels_df.groupby(['name']).max()
hotels_max.head()
```

```python
#Filter to only the relevant fields
hotels_final =  hotels_max[['address', 'city', 'province', 'postalCode']]
hotels_final['avgRating'] = hotels_mean['reviews.rating']
hotels_final.head()
```

```python
#Remove hotels that have no review or have NAs
hotels_final.drop(hotels_final.loc[hotels_final['avgRating'] == 0].index, inplace=True)
hotels_final.dropna(inplace=True)
hotels_final
```

```python
#Add helper columns that contain the lengths of the zip codes and states to identify bad data
hotels_final['len_state'] = hotels_final['province'].apply(lambda x: len(x))
hotels_final['len_zip'] = hotels_final['postalCode'].apply(lambda x: len(x))
```

```python
#Remove the bad data and drop the helper columns
hotels_final.drop(hotels_final.loc[hotels_final['len_state'] > 2].index, inplace=True)
hotels_final.drop(hotels_final.loc[hotels_final['len_zip'] != 5].index, inplace=True)
hotels_final.drop(columns=['len_state', 'len_zip'], inplace=True)
```

```python
#open the connection to mongodb and call the database
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.test
```

```python
#Iterate through the remaining data and load them into the collection
for row in hotels_final.iterrows():
    db.test.insert_one(
        {
            "Listing Name":row[0],
            "Address":row[1]['address'],
            "City":row[1]['city'],
            "State":row[1]['province'],
            "Zip":row[1]['postalCode'],
            "Average Rating":row[1]['avgRating'],
            "Type":"Hotel"
        }
    )
```
