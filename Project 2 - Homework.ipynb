```python
import pandas as pd
from sqlalchemy import create_engine
import requests
import warnings
warnings.filterwarnings('ignore')
```

```python
f = "airbnb-listings.csv"
f_df = pd.read_csv(f, delimiter=';', encoding = 'latin-1')
f_df.head()
```

```python
f_df.columns
```

```python
n_df = f_df[[
    'Name','Street',
    'City', 'State', 'Zipcode','Property Type', 
    'Review Scores Rating']].copy()
n_df.head()
```

```python

n_df.rename(columns={'Zipcode':'Zip',
                     'Property Type':'Type',
                     'Number of Reviews':'Review Count',
                     'Street':'Address',
                     'Review Scores Rating':'Average Rating'}, 
                 inplace=True)
```

```python
d = n_df.dropna()
```

```python
d['State'] = d['State'].str.upper()
```

```python
d.State.unique()
#
```

```python
#d = d.set_index('Name')
```

```python
d['Ratings2'] = d['Average Rating'] * (5/100)
```

```python
d = d.drop(['Average Rating'],axis=1)
d.rename(columns={'Ratings2':'Average Rating'}, 
                 inplace=True)
d.head()
```

```python
database_path = "db5.sqlite"
engine2 = create_engine(f"sqlite:///{database_path}")
# Query All Records in the the Database
d.to_sql(name='airbnb', con=engine2, if_exists='append', index=False)
data = engine2.execute("SELECT * FROM airbnb")

#for record in data:
 #   print(record)
```
